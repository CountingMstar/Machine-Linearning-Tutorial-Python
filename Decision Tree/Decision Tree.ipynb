```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
# 문자를 숫자로 매핑
from sklearn.tree import DecisionTreeClassifier
#의사결정 트리 분류 모듈
```

```python
df = pd.read_csv('salaries.csv')
# csv파일을 불러옴

df.head()
# 맨 앞 5개만 출력
```

```python
inputs = df.drop('salary_more_then_100k', axis='columns')
#inputs = df[['company', 'job', 'degree']]
# 'salary_more_then_100k'열(columns)을 뺀 나머지

inputs.head(3)
```

```python
target = df['salary_more_then_100k']
#target = df.salary_more_then_100k
target
```

```python
encoder = LabelEncoder()
# 문자를 숫자로 매핑
```

```python
inputs['company_n'] = encoder.fit_transform(inputs['company'])
inputs['job_n'] = encoder.fit_transform(inputs['job'])
inputs['degree_n'] = encoder.fit_transform(inputs['degree'])
# fit_transform : 데이터를 피팅하고 라벨숫자로 변환
inputs
# google=2, abc pharma=0, facebook=1
```

```python
inputs_n = inputs.drop(['company', 'job', 'degree'], axis='columns')
inputs_n
```

```python
model = DecisionTreeClassifier()
#의사결정 트리 분류 모듈
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n, target)
# 실제값과 예측값이 잘 맞는지의 점수
```

```python
model.predict([[2, 0, 1]])
# google, business manager, masters의 연봉 예측
```
